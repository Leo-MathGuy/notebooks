# Dictionaries

• List of key:value pairs


• Like a dictionary, you enter the key, and it outputs the value

• Unordered

• Curly braces and colons make a dictionary.

```python
example = {"Key":"Value", "Key2":"Value2"}
```

```python
example
```

```python
example["Key"]
```

## List vs. Dictionary

### Dictonary:

• Value retrieved by key.

• Unordered, so it cannot be sorted, indexed or sliced

### List:

• Value retrieved by location(index)

• Can be sorted, indexed, and sliced

## Examples

### Where to Use?

```python
prices = {"Gallon_Milk":6.99, "Small_Milk":4.45, "Chocolate_bar":2.99}
```

```python
prices["Gallon_Milk"]
```

### Uses

```python
favorite_food = {"Leo":["Sushi", "Donuts", "More Sushi"], "Lucy":{"Chicken":["Chicken", "More Chicken", "Even More Chicken"], "Other":["Kibble"]}}
```

```python
favorite_food["Leo"]
```

```python
favorite_food["Leo"][2]
```

```python
favorite_food["Lucy"]["Chicken"]
```

```python
myfood = favorite_food["Leo"]
```

```python
myfood[0]
```

```python
myfood
```

### Flexibility

```python
c = favorite_food["Lucy"]["Chicken"]
x = c[0]
x.upper()
```

```python
favorite_food["Lucy"]["Chicken"][0].upper()
```

```python
favorite_food["Anna"] = ["Chocolate", "Coffee", "Ice-Cream"]
```

```python
favorite_food["Anna"]
```

```python
favorite_food["Lucy"]["Other"] = ["Water", "Kibble"]
```

```python
favorite_food["Lucy"]
```
