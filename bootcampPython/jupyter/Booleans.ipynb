# Booleans

## What are they?

Booleans are True/False variables

Used in logic statements

## Examples

```python
type(True)
```

```python
had_coffee = False
```

```python
if(not(had_coffee)):
    print("Drinking coffee...")
    print("Coffee drank")
    had_coffee = True
```

```python
had_coffee = True
if(not(had_coffee)):
    print("Drinking coffee...")
    print("Coffee drank")
    had_coffee = True
else:
    print("I already had coffee")
```
