# Guessing Game Challenge

Let's use `while` loops to create a guessing game.

The Challenge:

Write a program that picks a random integer from 1 to 100, and has players guess the number. The rules are:

1. If a player's guess is less than 1 or greater than 100, say "OUT OF BOUNDS"
2. On a player's first turn, if their guess is
 * within 10 of the number, return "WARM!"
 * further than 10 away from the number, return "COLD!"
3. On all subsequent turns, if a guess is 
 * closer to the number than the previous guess return "WARMER!"
 * farther from the number than the previous guess, return "COLDER!"
4. When the player's guess equals the number, tell them they've guessed correctly *and* how many guesses it took!

You can try this from scratch, or follow the steps outlined below. A separate Solution notebook has been provided. Good luck!


```python
#Made by Leonid Iudakov 2021

#Import the library
from random import *

#Make variables
num = randint(1, 100)
i = 0

#Main Loop
while True:
    try: #Get guess
        guess = int(input("Please enter your guess from 1 to 100: "))
    except ValueError: #Show error if not integer
        print("ONLY DIGITS!")
        continue
    
    #Increase guess counter
    i+=1
    
    if guess < 1 or guess > 100: #Out of bounds error
        print("OUT OF BOUNDS")
        i-=1
        continue
        
    elif guess == num: #If right, show guesses and congrats
        print(f"You got it! Guesses: {i}!")
        
    elif abs(num - guess) <= 10 and i == 1: #1st guess warm? (In range of 10?)
        print("WARM!")
        last = abs(num - guess)
        continue
        
    elif i == 1: #1st guess cold?
        print("COLD!")
        last = abs(num - guess)
        continue
    
    elif i != 1 and abs(num - guess) < last:
        print("WARMER!") #Warmer?
        last = abs(num - guess)
        continue
        
    elif i != 1 and abs(num - guess) > last:
        print("COLDER!") #Colder?
        last = abs(num - guess)
        continue
    
    elif i != 1 and abs(num - guess) == last:
        print("SAME!") #Same?
        continue
```
