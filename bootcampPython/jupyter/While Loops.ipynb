# While Loops

## What are they?

While loops are loops that continiue looping while a condition is true.

For example, while pool level is less than 7 feet, continiue pumping water.

## Syntax

```python
#while *condition*:
    #code
#else:
    #code
```

## Examples

```python
#Example
x = 0
while x < 5: #while x is less than 5,
    print(x) #print x,
    x+=1     #and increase x by 1
```

```python
t = 10
while t >= 0:
    print(f"T-{t}")
    t-=1
else:
    print("Liftoff!") # If the condition isnt true, do this.
```

## Break, Continue, Pass

Break: Breaks out of current loop<br>
Continue: Goes to the start of current loop<br>
Pass: Do nothing

```python
while True:
    print("this will print")
    break
    print("this won't")
```

```python
x = 0
while x < 5:
    x += 1
    if x % 2:
        continue
    print(x)
```

```python
s = "Hello, World"
for x in s:
    if x == 'l':
        continue
    print(x)
```

```python
x = 0
if x < 5:
    pass #Placeholder for no syntax error
```
