# If, Else, And Elif

## Explanation

If, Else, and Elif **Control Flow**, meaning they can make some part of the code not run under some logic condition

For example, if there is rain outside, grab an umbrella.

In Python, Indentation(Whitespace) is very important to the syntax of if, else, and eli statements.

## Syntax

```python
#if *condition*:
    #some code
#elif *other condition*:
    #some other code
#else:
    #some other code
```

## Example

```python
#Example
#if condition1:
    #code1           If condition 1 is true, do code 1
#elif condition2:
    #code2           If condition 1 is false but condition 2 is true, do code 2 NOTE: There can be as much
#else:               elif statements as you want.
    #code3           If all conditions are false, run code 3
```

```python
if True:
    print("This will always run")
```

```python
if False:
    print("This will never run")
else:
    print("But this will")
```

```python
if False:
    print("This will not run")
elif True:
    print("This will always run")
```

```python
had_coffee = False
```

```python
if not(had_coffee):
    print("Drinking Coffee...\nDone!")
    had_coffee = True
else:
    print("I already had coffee")
```
