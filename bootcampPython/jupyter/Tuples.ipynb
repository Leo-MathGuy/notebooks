# Tuples

## What are they?

Tuples are lists, but are immutable and cannot be changed

They use parentheses: ()

Once a tuple is assigned the items, you cannot change one item without re-writing the whole tuple

## Example:

### How like lists

```python
t = (1, 2, 3)
l = [1, 2, 3]
```

```python
type(t)
```

```python
type(l)
```

```python
len(l)
```
