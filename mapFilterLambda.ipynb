# Map/Filter

```python
def square(x):
    return x**2
```

```python
nums = list(range(1, 11))
print(nums)
```

## Old Method - For Loop

```python
for i in nums:
    s = square(i)
    print("The square root of {} is {}".format(i, s))
```

```python
map(square, nums)
```

```python
for item in map(square, nums):
    print(item)
```

```python
list(map(square, nums))
```

```python
def oddfirst(n):
    if len(n)%2 == 1:
        return("EVEN")
    else:
        return n[0]
```

```python
def printlist(l):
    for i in l:
        print(i)
```

```python
names = ["Leo", "Anna", "Lucy", "Bob"]
```

```python
first = list(map(oddfirst, names))
```

```python
printlist(first)
```

## Filter

```python
def is_even(n):
    return not n % 2
```

```python
nums = [2, 25, 29, 92, 20, 8, 13]
```

```python
def find_even(l):
    return list(filter(is_even, l))
```

```python
printlist(find_even(nums))
```

## Lambda

```python
def square(num):
    result = num ** 2
    return result
```

```python
square(3)
```

```python
def square(num):
    return num ** 2
```

```python
square(3)
```

```python
def square(num): return num ** 2
```

```python
square(3)
```

```python
square = lambda num: num ** 2
```

```python
square(3)
```

### Lambda with map()

```python
nums = list(range(1, 11))
```

```python
list(map(lambda num: num ** 2, nums))
```

### Lambda with filter()

```python
list(filter(lambda num: not num % 2, nums))
```
